```python
movies = movies.merge(credits, on='title')
movies.shape
```

```python
movies.head()
```

```python
movies = movies[['movie_id','original_title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies.rename(columns = {'original_title': 'title'}, inplace = True)
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

```python
def convert2(text):
    l = []
    count = 5
    for i in ast.literal_eval(text):
        if count > 0:
            l.append(i['name'])
        count-=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert2)
movies.head(2)
```

```python
def fetch_dirprod(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
        if i['job'] == 'Executive Producer':
            if i['name'] not in L:
                L.append(i['name'])
        if i['job'] == 'Producer':
            if i['name'] not in L:
                L.append(i['name'])
        
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_dirprod)
movies.head(2)
```

```python
movies.sample(5)
```

```python
def collapse(l):
    l1 = []
    for i in l:
        l1.append(i.replace(" ",""))
    return l1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']
```

```python
new = movies.drop(columns = ['overview', 'genres', 'keywords'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
new.head(2)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
